## 废物

```python
pool = DataFrame([['sh.600096',0,1000000,1000000]],columns=['code','amount','cash','assets'])
def rebalancing(sig):
    global pool
    
    # 将alpha为正的股票平仓
    sell = sig[sig['alpha_x']>0]
    selling = pd.merge(pool,sell,on='code')
    selling['balancing'] = selling['amount'] * (selling['profit']+1) * (-1)
    selling['balancing_amount'] = selling['amount'] * (-1)
    selling['cash'] = selling['cash'] - (selling['balancing'].sum())
    selling['balancing_return'] = (selling['amount'] * selling['profit']).sum()
    # selling['balancing_return_rate'] = selling['balancing_return'] / selling['assets']
    selling['assets'] = selling['assets'] + selling['balancing_return']
    selling['amount'] = 0
    selling['tag'] = 'AS'
    
    # 对alpha为负的股票调仓
    buy = sig[sig['alpha_x']<0]
    buying = pd.merge(pool,buy,how='right',on='code')
    buying['weight'] = buying['profit'] / (buying['profit'].sum())
    buying['E_amount'] = buying['assets'] * buying['weight'] / buying['close'] // 100*100
    # return buying
    # 部分补仓或平仓
    buying_in = pd.merge(pool,buying,on=['code','amount','cash','assets'])
    # 部分平仓
    buying_in_sell = buying_in[buying_in['E_amount'].values<buying_in['amount'].values]
    buying_in_sell['balancing_amount'] = buying['E_amount'] - buying_in_sell['amount']
    buying_in_sell['balancing'] = buying_in_sell['balancing_amount']*buying_in_sell['close']
    buying_in_sell['cash'] = buying_in_sell['cash'] - (buying_in_sell['balancing'].sum())
    buying_in_sell['balancing_return'] = (buying_in_sell['balancing_amount'] * buying_in_sell['profit']).sum()
    # buying_in_sell['balancing_return_rate'] = buying_in_sell['balancing_return'] / buying_in_sell['assets']
    buying_in_sell['assets'] = buying_in_sell['assets'] + buying_in_sell['balancing_return']
    buying_in_sell['amount'] = buying['E_amount']
    buying_in_sell['tag'] = 'PS'
    # 部分升仓
    buying_in_buy = buying_in[buying_in['E_amount'].values>buying_in['amount'].values]
    buying_in_buy['balancing_amount'] = buying['E_amount'] - buying_in_buy['amount']
    buying_in_buy['balancing'] = buying_in_buy['balancing_amount']*buying_in_buy['close']
    buying_in_buy['cash'] = buying_in_buy['cash'] - (buying_in_buy['balancing'].sum())
    buying_in_buy['amount'] = buying['E_amount']
    buying_in_buy['tag'] = 'PB'
    
    
    # 未持有按权重补仓
    buying_not_in = buying[buying['code'].values!=buying_in['code'].values]
    buying_not_in['balancing'] = buying['E_amount']*buying_not_in['close']
    buying_not_in['balancing_amount'] = buying['E_amount']
    buying_not_in['cash'] = buying_not_in['cash'] - (buying_not_in['balancing'].sum())
    buying_not_in['amount'] = buying['E_amount']
    buying_not_in['tag'] = 'AB'
    
    sell_total = pd.concat([selling,buying_in_sell])
    buy_total = pd.concat([buying_in_buy,buying_not_in])
    total = pd.concat([sell_total,buy_total])
    pool = pd.concat([pool,total])
    return total
pool
```
